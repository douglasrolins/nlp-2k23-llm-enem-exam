# Importar Bibliotecas Básicas e Definir Funções Básicas

```python
## Importar bibliotecas básicas

import numpy as np
import pandas as pd
import time
import os
import math
from sklearn.metrics import accuracy_score
```

```python
## Definir funções básicas

# Função para gerar as métricas
def get_metrics(trues,preds):

  # Converter os DataFrames em listas
  y_true_list = trues.tolist()
  y_pred_list = preds.tolist()

  # Calcular a taxa de acertos
  taxa_acertos = accuracy_score(y_true_list, y_pred_list)

  # Calcular a taxa de erro
  taxa_erro = 1 - taxa_acertos

  # Obter o número total de questões
  num_questoes = len(y_true_list)

  # Calcular o número de acertos e o número de erros
  num_acertos = int(taxa_acertos * num_questoes)
  num_erros = num_questoes - num_acertos

  # Resultados
  print(f'Número de Acertos: {num_acertos}')
  print(f'Número de Erros: {num_erros}')
  print(f'Taxa de Acerto: {taxa_acertos * 100:.2f}%')
  print(f'Taxa de Erro: {taxa_erro * 100:.2f}%')
```

```python
# Instalar sumy
!pip install sumy

# Importar bibliotecas
from sumy.parsers.plaintext import PlaintextParser
from sumy.summarizers.text_rank import TextRankSummarizer
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.tokenizers import Tokenizer
from sumy.utils import get_stop_words
import nltk
nltk.download('punkt')
import re
```

```python
# Funções para Pré-Processamento

# Remover números
def remove_numbers_in_text(text):
  return re.sub(r'\d+', '', text)

# Remover caracteres especiais
def remove_special_chars_in_text(text):
  return re.sub(r'[^\w\s]', '', text)

#ajustar campos em branco
def remove_blank(df):
  df = df.fillna("",inplace=True)
  return df

# Converter texto para minísculas
def minusculo(df):
  for col in df.columns:
    if df[col].dtype == 'object':  # Verifica se a coluna contém texto (objeto).
      df[col] = df[col].str.lower()
  return df

# stop words
stop_words = get_stop_words('portuguese')

# definir a função de sumarização
def sum_text(texto,percent):
  # obtém as stop words para o idioma inglês
  stop_words = get_stop_words("portuguese")
  # realiza a sumarização do texto utilizando a biblioteca sumy
  parser = PlaintextParser.from_string(texto, Tokenizer("english"))
  # definir tamanho máximo
  sentence_count = len(parser.document.sentences)
  summary_size = math.ceil(sentence_count * percent)

  #summarizer = LsaSummarizer()
  #summarizer = TextRankSummarizer()
  summarizer = LexRankSummarizer()

  summarizer.stop_words = stop_words # define as stop words a serem removidas
  summary = summarizer(parser.document, summary_size)
  resumo = ""
  for sentence in summary:
    resumo += str(sentence) + " "
  # retorna o resumo do texto
  return resumo
```

```python
# Sumarizar a sentenção e remover stop-words
percent_summarize = 0.01
dfTrain['content'] = dfTrain['content'].apply(lambda x: sum_text(x, percent_summarize))
```

```python
# Converter para minusculo
dfTrain = minusculo(dfTrain)
dfTest = minusculo(dfTest)
```

# Método Inicial - Aleatório

Método inicial para uma base de referência. Qualquer solução que retorne resultado igual ou menor que este método inicial será descartada.

```python
# 1º método aleatório: percorre cada questão e escolhe uma alternativa aleatoriamente
import random

def gera_respostas(df_questoes):

  # Cria um novo dataframe para armazenar as respostas
  df_respostas = pd.DataFrame(columns=['id', 'prediction'])

  # Itera sobre as linhas do dataframe de questões
  for idx, row in df_questoes.iterrows():
    # Gera uma alternativa aleatória entre A, B, C, D, E
    resposta_aleatoria = random.choice(['A', 'B', 'C', 'D', 'E'])

    # Adiciona a resposta ao dataframe de respostas
    df_respostas = df_respostas.append({'id': idx, 'prediction': resposta_aleatoria}, ignore_index=True)

  return df_respostas
```

```python
# 2º método aleatório: cria um array no tamanho da quantidade de questões distribuindo aleatoriamente as 5 alternativas
import random

def gera_respostas(df_questoes):

  # Defina o tamanho do array (substitua X pelo tamanho desejado)
  tamanho_array = len(df_questoes)

  # Gere o array com valores aleatórios entre A, B, C, D e E
  valores_possiveis = np.array(['A', 'B', 'C', 'D', 'E'])
  array_respostas = np.random.choice(valores_possiveis, tamanho_array)

  # Crie um novo DataFrame com as respostas
  df_respostas = pd.DataFrame({'id': df_questoes.index, 'prediction': array_respostas})

  return df_respostas
```

```python
# Carregar modelo word2vec
from gensim.models import KeyedVectors
model = KeyedVectors.load_word2vec_format('cbow_s300.txt')
```

```python
# Carregar tokenizador
from nltk import TweetTokenizer

tokenizer = TweetTokenizer()
```

```python
# from logging import exception

# Função embedding para cada palavra. Se a palavra existir, gerar embedding da palavra. Senão gerar embedding com valores zero nas dimensões
def gerarEmbedding (word):
  try:
    embedding = model[word]
  except:
    embedding = np.zeros(model.vector_size)

  return embedding
```

```python
# Função para tokenizar, gerar embeddings com word2vec
# Retorna os embeddings prompt/alternatives e content

def gera_embeddings_word2vec(df,model):

  all_questions_emb = []
  all_alternatives_emb = []

  for index, row in df.iterrows():
    prompt = str(row['prompt'])
    content = str(row['content'])
    question = (content+ " " +prompt)

    #tokens_prompt = tokenizer.tokenize(prompt)
    #tokens_content = tokenizer.tokenize(content)
    tokens_question = tokenizer.tokenize(question)

    #embedding_prompt = [gerarEmbedding(t) for t in tokens_prompt]
    #embedding_content = [gerarEmbedding(t) for t in tokens_content]
    embedding_question = [gerarEmbedding(t) for t in tokens_question]

    #media_embedding_prompt = np.mean(embedding_prompt,axis=0).astype(np.float32)
    #media_embedding_content = np.mean(embedding_content,axis=0).astype(np.float32)
    media_embedding_question = np.mean(embedding_question,axis=0).astype(np.float32)

    alternatives = row[['A', 'B', 'C', 'D', 'E']].values
    alternatives_emb = []
    for alt in alternatives:
      tokens_alt = tokenizer.tokenize(alt)
      embedding_alt = [gerarEmbedding(t) for t in tokens_alt]
      media_embedding_alt = np.mean(embedding_alt,axis=0).astype(np.float32)
      alternatives_emb.append(media_embedding_alt)

    # Adiciona os embeddings no array com todos os embeddings
    all_questions_emb.append(media_embedding_question)
    all_alternatives_emb.append(alternatives_emb)

  return all_questions_emb, all_alternatives_emb
```

```python
# Função para realizar o cálculo da similaridade entre as questões e as alternativas
# Retorna uma lista com o ID de cada questão e a alternativa com maior similaridade

from sklearn.metrics.pairwise import cosine_similarity

def predicao_similaridade(questions_embeddings_all,alternatives_embeddings_all):

  predicoes = []
  index_alternatives = ["A", "B", "C", "D", "E"] # indice das questões
  id = 0
  for q, a in zip(questions_embeddings_all, alternatives_embeddings_all):

    # Certificar-se de que as entradas são arrays 2D
    q = np.array(q).reshape(1, -1)

    # Calcular a similaridade (dot) entre a questão e cada uma das 5 alternativas
    scores = cosine_similarity(q, a)

    # Combinar as alternativas e a similaridade com a questão. Incluir ainda o índice da alternativa (letra)
    score_alternatives = list(zip(index_alternatives, a, scores.flatten()))

    # Ordenar por ordem decrescente de score
    score_alternatives = sorted(score_alternatives, key=lambda x: x[2], reverse=True)

    # Selecionar a alternativa com o maior score como a resposta prevista
    resposta_prevista = score_alternatives[0][0]

    # Armazenar o ID da questão e a resposta prevista na lista de predições
    predicoes.append((id, resposta_prevista))
    id = id + 1

  return predicoes
```

```python
## Fazer predição na base Train

start = time.time()

# Gerar embeddings
emb_q, emb_a = gera_embeddings_word2vec(dfTrain,model)

# Realizar predição pela similaridade
predicoes = predicao_similaridade(emb_q,emb_a)

end = time.time()

print('Tempo para processamento: %2.fs \n' % (end-start),'Quantidade de questões: ', len(predicoes))

preds = pd.DataFrame(predicoes)
get_metrics(dfTrain['answer'],preds[1])
```

```python
## Fazer predição na base Test (para submissão na competição)

start = time.time()

# Gerar embeddings
emb_q, emb_a = gera_embeddings_word2vec(dfTest,model)

# Realizar predição pela similaridade
predicoes = predicao_similaridade(emb_q,emb_a)

end = time.time()

print('Tempo para processamento: %2.fs \n' % (end-start),'Quantidade de questões: ', len(predicoes))

preds = pd.DataFrame(predicoes)
```

```python
## Converter para arquivo CSV para submeter na competição
preds = pd.DataFrame(predicoes)
preds = preds.rename(columns={0: 'id',1:'prediction'})
url_file_result = 'predict.csv'
preds.to_csv(url_file_result, header=True, index=False)
```

### Realizar Fine-Tuning Modelo Sentence-Transformers

```python
# Importar bibliotecas necessárias
from torch.utils.data import DataLoader
from sentence_transformers import InputExample, losses, evaluation
```

```python
# Parâmetros do treinamento
#@title Setar Parâmetros
#@markdown Parâmetros do treinamento:
num_epochs = 5 #@param ["1", "3", "5", "10", "20", "40", "50", "60", "80", "100", "200", "1000"] {type:"raw"}
batch_size = 8 #@param ["2", "4", "8", "16", "32", "64", "128"] {type:"raw"}
learning_rate = 2e-5 #@param ["1e-5", "2e-5", "3e-5", "4e-5", "5e-5", "1e-4"] {type:"raw"}
#(default 2e-5)
scheduler = 'ConstantLR' #@param ['ConstantLR', 'WarmupConstant', 'WarmupLinear', 'WarmupCosine', 'WarmupCosineWithHardRestarts']
save_best_model_ft = True #@param {type: 'boolean'}
```

```python
# Diretório para salvar modelo
base_path = '/content/drive/MyDrive/Colab Notebooks/'
path_save_model = str(base_path) + 'models/' + str(model_pre) + '_' + str(num_epochs) +'_'+str(batch_size)+'/'
```

```python
# Método para gerar examples para treinamento
letra_para_numero = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}
def gera_input_examples(df):
  examples = []
  for d in df.index:
    content = df['content'][d]
    prompt = df['prompt'][d]

    question = f"{content} {prompt}"

    alternatives = dfTrain.loc[d, ['A', 'B', 'C', 'D', 'E']].tolist()
    answer = df['answer'][d]

    # Substituto a letra de answer por label inteiro
    label_index = letra_para_numero[answer]

    # Criação da lista de rótulos
    labels = [0] * 5
    labels[label_index] = 1

    for alternative, label in zip(alternatives,labels):
      examples.append(InputExample(texts=[question, alternative], label=int(label)))
  return examples
```

```python
# Dividir os dados em treinamento e validação
from sklearn.model_selection import train_test_split

# porcentagem desejada para treino (restante será validação)
percent_treino = 0.7

# Use a função train_test_split para dividir o DataFrame
df_treino, df_validacao = train_test_split(dfTrain, test_size=1-percent_treino, random_state=42)
```

```python
# Gera exemplos de treinamento e validação

train_examples = gera_input_examples(df_treino)
valid_examples = gera_input_examples(df_validacao)
```

```python
 # Definindo o dataloader e parâmetros de treinamento
train_batch_size = batch_size
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=train_batch_size)
steps_per_epoch = len(train_dataloader) // train_batch_size
num_evaluator_steps = len(train_dataloader)
num_total_steps = steps_per_epoch * num_epochs
warmup_steps = num_total_steps
```

```python
# Definindo a função de perda

#ConstrastiveLoss
train_loss = losses.ContrastiveLoss(model=model_st)

#OnlineConstrativeLoss
#margin = 0.5
#distance_metric = losses.SiameseDistanceMetric.COSINE_DISTANCE
#train_loss = losses.OnlineContrastiveLoss(model=model, distance_metric=distance_metric, margin=margin)
```

```python
# Definir evaluator
#evaluator = evaluation.EmbeddingSimilarityEvaluator.from_input_examples(valid_examples, name='valid_examples', show_progress_bar=True)

evaluator = evaluation.BinaryClassificationEvaluator.from_input_examples(valid_examples, name='valid_examples', show_progress_bar=True)
```

```python
# Ajuste no log para exibir as informações no momento do treinamento
import logging
from sentence_transformers import LoggingHandler
logging.basicConfig(format='%(asctime)s - %(message)s',
                    datefmt='%Y-%m-%d %H:%M:%S',
                    level=logging.INFO,
                    force=True,
                    handlers=[LoggingHandler()])
```

```python
# Realizar treinamento

start = time.time()

model_st.fit(train_objectives=[(train_dataloader, train_loss)],
          evaluator=evaluator,
          epochs=num_epochs,
          steps_per_epoch=steps_per_epoch,
          evaluation_steps = num_evaluator_steps,
          show_progress_bar=True,
          save_best_model=save_best_model_ft,
          scheduler = scheduler,
          warmup_steps=warmup_steps,
          optimizer_params={'lr': learning_rate},
          output_path = path_save_model)
end = time.time()

print('- Tempo para fine-tuning: %2.fs \n' % (end-start))
```

### Realizar Inferência na Base Test

```python
# Função para fazer a predição para todas as questões do DataFrame.
# Passa por cada questão e gerar os embedddings, após calcular a similaridade entre o prompt/content e alternativas e
# selecionar a alternativa com maior similaridade.

def fazer_predicao(df,include_content=False):
  predicoes = []  # Inicializa uma lista para armazenar as predições
  index_alternatives = ["A", "B", "C", "D", "E"] # indice das questões

  for index, row in df.iterrows():
    question = str(row['prompt'])
    content = str(row['content'])
    alternatives = row[['A', 'B', 'C', 'D', 'E']].values

    # Gerar embeddings da questão e das 5 alternativas
    if (include_content):
      question_emb = model_st.encode(content + " " + question, show_progress_bar = False)
    else:
      question_emb = model_st.encode(question, show_progress_bar = False)
    alternatives_emb = model_st.encode(alternatives, show_progress_bar = False)

    # Calcular a similaridade (dot) entre a questão e cada uma das 5 alternativas
    scores = util.dot_score(question_emb, alternatives_emb)[0].cpu().tolist()

    # Combinar as alternativas e a similaridade com a questão. Incluir ainda o índice da alternativa (letra)
    score_alternatives = list(zip(index_alternatives, alternatives, scores))

    # Ordenar por ordem decrescente de score
    score_alternatives = sorted(score_alternatives, key=lambda x: x[2], reverse=True)

    # Selecionar a alternativa com o maior score como a resposta prevista
    resposta_prevista = score_alternatives[0][0]

    # Obter o ID da questão
    question_id = row['id']

    # Armazenar o ID da questão e a resposta prevista na lista de predições
    predicoes.append((question_id, resposta_prevista))

  return predicoes
```

```python
# Gerar embeddings da base de teste
start = time.time()

predicoes = fazer_predicao(dfTest,include_content=True)

end = time.time()

print('Tempo para processamento: %2.fs \n' % (end-start),'Quantidade de questões: ', len(predicoes))
```

```python
## Converter para arquivo CSV para submeter na competição
preds = pd.DataFrame(predicoes)
preds = preds.rename(columns={0: 'id',1:'prediction'})
url_file_result = 'predict.csv'
preds.to_csv(url_file_result, header=True, index=False)
```

```python
if realizar_treinamento:

  # Método para computar métricas
  accuracy = evaluate.load("accuracy")
  def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)

  # Método para processar tokenização dos dados e gerar os exemplos de treinamento
  # Retorna duas sentenças tokenizadas, uma com um contexto (5 vezes) e a outra com prompt + cada alternativa

  respostas_alternativas = ["A", "B", "C", "D", "E"]

  ## old: função do tutorial multiple-choice do hf
  def preprocess_function_old(examples):
    first_sentences = [[context] * 5 for context in examples["content"]]
    question_headers = examples["prompt"]
    second_sentences = [
      [f"{header} {examples[end][i]}" for end in respostas_alternativas] for i, header in enumerate(question_headers)
    ]

    first_sentences = sum(first_sentences, [])
    second_sentences = sum(second_sentences, [])

    tokenized_examples = tokenizer(first_sentences, second_sentences, truncation=True)
    return {k: [v[i : i + 5] for i in range(0, len(v), 5)] for k, v in tokenized_examples.items()}

  ## new: ajustado para 5 alternativas, bem como para incluir o contexto e prompt na primeira sentença e as alternativas na segunda sentença
  def preprocess_function(examples):
    first_sentences = [[f'{context} {prompt}'] * 5 for context, prompt in zip(examples["content"],examples["prompt"])]

    second_sentences = [
      [f"{examples[alt][i]}" for alt in respostas_alternativas] for i, header in enumerate(examples["prompt"])
    ]

    first_sentences = sum(first_sentences, [])
    second_sentences = sum(second_sentences, [])

    tokenized_examples = tokenizer(first_sentences, second_sentences, truncation=True)
    return {k: [v[i : i + 5] for i in range(0, len(v), 5)] for k, v in tokenized_examples.items()}
```

```python
if realizar_treinamento:

  # Ordenação dos dados de multipla escolha

  @dataclass
  class DataCollatorForMultipleChoice:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
      label_name = "label" if "label" in features[0].keys() else "labels"
      labels = [feature.pop(label_name) for feature in features]
      batch_size = len(features)
      num_choices = len(features[0]["input_ids"])
      flattened_features = [
          [{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features
      ]
      flattened_features = sum(flattened_features, [])

      batch = self.tokenizer.pad(
        flattened_features,
        padding=self.padding,
        max_length=self.max_length,
        pad_to_multiple_of=self.pad_to_multiple_of,
        return_tensors="pt",
      )

      batch = {k: v.view(batch_size, num_choices, -1) for k, v in batch.items()}
      batch["labels"] = torch.tensor(labels, dtype=torch.int64)
      return batch
```

```python
if realizar_inferencia:

  # qtd máxima de tokens
  max_length = model.config.max_position_embeddings

  # Função para predizer a resposta correta para uma questão do DataFrame
  def predict(row):

    total_text = row['content'] + " " + row['prompt']

    # Truncar o texto se for muito longo
    #total_text = total_text[:max_length] # pegar primeiros tokens
    total_text = total_text[-max_length:] # pegar últimos tokens

    alternatives = row[['A', 'B', 'C', 'D', 'E']].values.tolist()

    inputs = tokenizer([[total_text, alternative] for alternative in alternatives], return_tensors="pt", padding=True)
    #inputs = tokenizer([[total_text, alternative] for alternative in alternatives], return_tensors="pt", padding=True, truncation=True, max_length=max_length)
    inputs = inputs.to(device)

    labels = torch.tensor(0).unsqueeze(0)
    labels = labels.to(device)

    outputs = model(**{k: v.unsqueeze(0) for k, v in inputs.items()}, labels=labels)
    logits = outputs.logits

    predicted_class_index = logits.argmax().item()
    predicted_class = chr(ord("A") + predicted_class_index)
    return predicted_class
```

```python
# Extrair letra correta com Regex

def extrair_letra_correta(resposta):

  # Divide a resposta em linhas
  linhas = resposta.split('\n')

  # Procura por padrões que indicam a letra correta na última linha
  padroes = [
      r'Resposta correta: ([A-E])',
      r'A alternativa correta é a ([A-E])',
      r'A resposta correta é: [A-E]',
      r'a alternativa correta é a opção [A-E]',
  ]

  for padrao in padroes:
    match = re.search(padrao, linhas[-1])  # A última linha
    if match and len(match.groups()) > 0:
      return match.group(1)  # Retorna o grupo correspondente à letra (A-E)

  # Retornar null caso não encontre o padrão definido (caso ocorra, verificar e incluir o novo padrão gerado pelo modelo)
  return 'null'
```

```python
# Extrair a letra correta pelo método regex

letras_respostas = []
id = 0
for r in respostas:
  letras_respostas.append((id,extrair_letra_correta(r)))
  id = id + 1
```

```python
# Função Regex para extrair letra correta da resposta gerada pelo modelo
import re

def extrair_letra_correta(resposta):

  # Divide a resposta em linhas
  linhas = resposta.split('\n')

  # Padrões que indicam a letra correta
  padroes = [
      #r'\((A|B|C|D|E)\)',  # Padrão para letras entre parênteses
      r'Resposta: ([A-E])',  # Padrão para "Resposta: A"
      r'A alternativa correta é a ([A-E])',  # Padrão para "A alternativa correta é a: A"
      r'a alternativa correta é a letra (A|B|C|D|E)',  # Padrão "a alternativa correta é a letra B"
      r'Resposta: ([A-E])',  # Padrão para "Resposta: E",
      r'alternativa correta é a (A|B|C|D|E)',  # Padrão "alternativa correta é a B"
      r'alternativa correta é (A|B|C|D|E)',  # Padrão "alternativa correta é B"
      r'correta é a alternativa (A|B|C|D|E)', #correta é a alternativa A
      r'Alternativa correta: (A|B|C|D|E)', # Alternativa correta: D
      r'correta é a letra (A|B|C|D|E)', # correta é a letra D
      r'correta é a letra \((A|B|C|D|E)\)', # correta é a letra (D)
      r'correta é \((A|B|C|D|E)\)', # correta é (B)
      r'alternativa correta é a \((A|B|C|D|E)\)', # A alternativa correta é a (C)
      r'correta é a alternativa \((A|B|C|D|E)\)', # correta é a alternativa (C)
      r'correta: \((A|B|C|D|E)\)',  # Padrão para "correta: (D)"
      r'correta é: \((A|B|C|D|E)\)'  # Padrão para "correta é: (C)""
  ]

  for padrao in padroes:
    match = re.search(padrao, linhas[-1])  # A última linha
    if match and len(match.groups()) > 0:
      return match.group(1)  # Retorna o grupo correspondente à letra (A-E)

  # caso não encontre algum padrão retornar null
  return 'null'
```

```python
# Extrair a letra correta pelo método regex
letras_respostas = []
id = 0
for r in respostas:
  letras_respostas.append((id,extrair_letra_correta(r)))
  id = id + 1
```
